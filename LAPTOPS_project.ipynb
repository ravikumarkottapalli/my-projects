```python
url2="https://www.flipkart.com/search?q=under+50000+laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=2"
page=requests.get(url2)
page

soup =BeautifulSoup(page.text)
soup

Brand1 = soup.find_all("div", class_="_4rR01T")

Brand = []
for Brand1 in soup.find_all("div", class_="_4rR01T"):
    Brand.append(Brand1.text.split()[0])
Brand


Or_Price = soup.find_all("div",class_="_3I9_wc _27UcVY")
Or_Price[0].text
Or_Price1 = []
for i in Or_Price:
    Or_Price1.append(i.text.split("₹")[1])

price1  = []
for i in soup.find_all("div",class_="_2kHMtA"):
    price= i.find("div",class_="_30jeq3 _1_WHN1")
    #print(price.text)
    price1.append(price.text.split()[0])

Discount1 =[]
for i in soup.find_all("div",class_="_2kHMtA"):
    Discount = i.find("div",class_="_3Ay6Sb")
    #print(Discount.text)
    Discount1.append(Discount.text.split()[0])

Processer1 = []
for i in soup.find_all("div",class_="_2kHMtA"):
    model_name = i.find("li",class_="rgWa7D")
    #print(model_name.text)
    Processer1.append( model_name.text.split()[0])

Ratings1 = []
for i in soup.find_all("div",class_="_2kHMtA"):
    cc = i.find("div",class_="_3LWZlK")
    if cc:
        Ratings1.append(cc.text)
    else:
        Ratings1.append(np.nan)

Reviews1 = []
for i in soup.find_all("div",class_="_2kHMtA"):
    aa = i.find("span",class_="_2_R_DZ")
    if aa:
        a =re.findall('\d+\sReviews',aa.text)
        Reviews1.append(a[0])
    else:
        Reviews1.append(np.nan)

ssd =[]
b = soup.find_all("div",class_ ="_4rR01T")
for i in b:
    a=i.text
    #print(a)
    SSD = re.findall("\/(\d+)\sGB\s\SSD",a)
    if len(SSD) == 1:
        ssd.append(SSD[0])
    else:
        ssd.append("1024")



data1 = {"Brand_Name":Brand,"Original_Price": Or_Price1,"Price":price1,"Discount_Percentage":Discount1,"Processer":Processer1,"ROM":ssd,"Ratings": Ratings1,"Reviews": Reviews}
#data1 
    
df1 = pd.DataFrame(data1)
df1
#df.to_csv("LAPTOPS_1.csv")
  
```
